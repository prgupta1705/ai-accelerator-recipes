### Trend Identification in Time Series Datasets

Identifying trends or patterns in a Time Series related datasets is of prime importance. This use case helps us to identify the trend/patterns in Time series datasets. It uses data_analysis module's trends() method whch wraps some standard Python modules.

### Dataset Information

The dataset used in the use case, is taken from the public domain and contains information about the data for individual stocks, labelled by their stock ticker name.

There are total 7 features in the dataset and they are listed below :-

Date - in format: yy-mm-dd

Open - price of the stock at market open (this is NYSE data so all in USD)

High - Highest price reached in the day

Low, Close - Lowest price reached in the day

Volume - Number of shares traded

Name - the stock's ticker name

```python
import pandas as pd
import numpy as np
from data_analysis import trends
```

```python
data = pd.read_csv('AMZN_2006-01-01_to_2018-01-01.csv')
```

```python
trends('Open', data, 'Date')
```

```python
trends('Close', data, 'Date')
```

```python
trends('High', data, 'Date')
```

```python
trends('Low', data, 'Date')
```

```python
trends('Volume', data, 'Date')
```
